Dataset Overview
This dataset contains health-related information collected from individuals, covering various aspects such as general health, physical and mental health, and medical history. The dataset includes both numerical and categorical features.

Sex: Categorical variable indicating gender (Male, Female).

GeneralHealth: A numeric rating of general health on a scale from 0 (worst) to 4 (best).

PhysicalHealthDays & MentalHealthDays: Number of days in the past month with poor physical or mental health.

PhysicalActivities: Indicates whether an individual engages in physical activities (Yes, No).

SleepHours: Number of hours of sleep per day.

Medical History: Includes conditions like heart disease, diabetes, asthma, arthritis, and more (Yes, No).

AgeCategory: Numeric representation of age groups (e.g., 18, 25, 30, ..., 80).

BMI: Body Mass Index (BMI), a numerical feature representing weight status.

Vaccination & Health Screening: Indicates whether an individual received flu or pneumonia vaccines and underwent medical tests (Yes, No).

```python
from sklearn.preprocessing import LabelEncoder

binary_columns = [
    'PhysicalActivities', 'HadHeartAttack', 'HadAngina', 'HadStroke', 'HadAsthma',
    'HadSkinCancer', 'HadCOPD', 'HadDepressiveDisorder', 'HadKidneyDisease',
    'HadArthritis', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty',
    'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing',
    'DifficultyErrands', 'ChestScan', 'AlcoholDrinkers', 'HIVTesting',
    'FluVaxLast12', 'PneumoVaxEver', 'HighRiskLastYear'
]

for col in binary_columns:
    if col in data.columns:
        data[col] = data[col].map({'Yes': 1, 'No': 0})

categorical_columns = ['Sex', 'HadDiabetes']
encoder = LabelEncoder()
for col in categorical_columns:
    if col in data.columns:
        data[col] = encoder.fit_transform(data[col])

print(data.head())
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data["Sex"] = encoder.fit_transform(data["Sex"])
```

```python
data['HadDiabetes'] = data['HadDiabetes'].replace({
    'No': 0,
    'Yes': 1,
    'Yes, but only during pregnancy (female)': 2,
    'No, pre-diabetes or borderline diabetes': 3
})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

data = data.replace({'Yes': 1, 'No': 0})  

x = data.drop(['HadHeartAttack'], axis=1)
y = data['HadHeartAttack']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

svm_model = SVC()
svm_model.fit(x_train, y_train)

y_pred = svm_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

y_pred = svm_model.predict(x_test)


accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Accuracy: {accuracy:.2f}%")
```
